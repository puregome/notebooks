# Counts of tweets collected by twiqs.nl crawler

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATAFILE = "twitter-counts.csv"
DATEFORMAT = "%Y%m%d-%H.out.gz"
```

```python
df = pd.read_csv(DATAFILE,header=None,index_col=0)
```

```python
N=-80

plt.figure(figsize=(20,8))
ax = plt.subplot(111)
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[6][N:],xdate=True,fmt="-",label="trackb")
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[5][N:],xdate=True,fmt="-",label="track")
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[1][N:],xdate=True,fmt="-",label="all")
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[4][N:],xdate=True,fmt="-",label="locations")
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[3][N:],xdate=True,fmt="-",label="follow")
ax.plot_date([datetime.strptime(d,DATEFORMAT) for d in df.index][N:],df[2][N:],xdate=True,fmt="-",label="dialect")
plt.legend()
plt.show()
```

## Per minute

```python
HOURFILE1 = "hours-20200701.csv"
HOURFILE2 = "hours-20200702.csv"
HOURFILE3 = "hours-20200703.csv"
HOURFORMAT = "%Y%m%d%H%M"
```

```python
dfHour1 = pd.read_csv(HOURFILE1,header=None,index_col=1)
dfHour2 = pd.read_csv(HOURFILE2,header=None,index_col=1)
dfHour3 = pd.read_csv(HOURFILE3,header=None,index_col=1)
```

```python
plt.figure(figsize=(20,8))
ax = plt.subplot(111)
ax.plot_date([datetime.strptime(str(d),HOURFORMAT) for d in dfHour1.index],dfHour1[0],xdate=True,fmt="-")
ax.plot_date([datetime.strptime(str(d),HOURFORMAT) for d in dfHour2.index],dfHour2[0],xdate=True,fmt="-")
ax.plot_date([datetime.strptime(str(d),HOURFORMAT) for d in dfHour3.index],dfHour3[0],xdate=True,fmt="-")
plt.show()
```
